```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

def draw_graph(graph):

    # extract nodes from graph
    nodes = set([n1 for n1, n2 in graph] + [n2 for n1, n2 in graph])

    # create networkx graph
    G=nx.Graph()

    # add nodes
    for node in nodes:
        G.add_node(node)

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # draw graph
    pos = nx.shell_layout(G) # to make circular layout for nodes
    nx.draw(G, pos, with_labels=True, node_color='blue', edge_color='pink', alpha=0.5, node_size=1000, font_size=16)

    # show graph
    plt.show()

# draw example
graph = [(20, 22),(21, 22),(22, 23), (23, 24),(24, 25), (25, 20)]
draw_graph(graph)

# plt.savefig("simple_path.png")
```

```python
#Create empty networkx graph
G=nx.Graph()

#Add nodes
G.add_nodes_from(["A","B","C","D","E","F","G", "H","I","J"])

#Specify edges
G.add_edges_from([("A","B"), ("A","C"), ("A","D"), ("A","J"), ("B","E"), ("B","F"),
("C","G"),("C","H"),("D","I")])

#Draw graph with matplotlib
nx.draw(G, nx.shell_layout(G), with_labels=True, node_color='blue', edge_color='pink', alpha=0.5, node_size=1000, font_size=16)
```

```python
# Different weights for different edges
G=nx.Graph()

G.add_weighted_edges_from([('a','b',0.6),('a','c',0.2),('c','d',1),('c','f',0.7)])
edgewidth = [ d['weight'] for (u,v,d) in G.edges(data=True)]

pos= nx.shell_layout(G)

nx.draw_networkx_nodes(G,pos,node_size=700)

nx.draw_networkx_edges(G,pos,width=edgewidth)

nx.draw_networkx_labels(G,pos)

plt.axis('off')
```

```python
# Different weights for different edges
G=nx.Graph()

G.add_weighted_edges_from([('a','b',0.6),('a','c',0.2),('c','d',0.1),('c','f',0.7)])

elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >0.5]
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] <=0.5]

pos=nx.shell_layout(G) # position for all nodes

# nodes
nx.draw_networkx_nodes(G,pos,node_size=700)

# edges
nx.draw_networkx_edges(G,pos,edgelist=elarge,width=6)
nx.draw_networkx_edges(G,pos,edgelist=esmall,width=6,alpha=0.5,edge_color='b',style='dashed')

# labels
nx.draw_networkx_labels(G,pos,font_size=20,font_family='sans-serif')

plt.axis('off')
```
