```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
import numpy as np
import matplotlib.cm as cm
```

```python
# Line plot
x=np.linspace(0,20, num=100) #num is number of samples
#print(x)
y1=np.sin(x)
y2=np.sin(x-np.pi)

# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

plt.plot(x,y1)
plt.plot(x,y2)
```

```python
# Format line and markers
plt.plot(x,
         y1,
         color='black',         
         linestyle='-',
         linewidth=2,
         marker='s',
         markersize=6,
         label='y1') # to appear in the legend

plt.plot(x,
         y2,
         color='gray',
         linestyle='--',
         linewidth=2,
         marker='^',
         markersize=6,
         label='y2') 

# can concatenate line color and style: plt.plot([1,2,3,4], [1,4,9,16], 'ro')

plt.title("My first plot",fontsize=16, fontweight='bold')
plt.xlabel('x-axis', fontsize=14, color='red')
plt.ylabel('y-axis',fontsize=14, color='red')

plt.legend(loc='upper right') #manual placing: bbox_to_anchor=(1, 0))
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')

# Format axis

## Set limits
plt.xlim(0,15)

## Show selected ticks only

plt.xticks([0,4,8,12])
plt.yticks([-1, 0, +1])

yticklabels=['-$1','$0','$1']

## Change tick labels -- suppose add $ sign before y-axis labels
# plt.yticks([-1,0,+1],[r'-$1',r'$0',r'$1'])
plt.yticks([-1,0,+1],yticklabels)

# Changing x and y axis position: these are called spines. 
## There are 4 spines- top, bottom, left, right

'''set_position(position)- set the position of the spine

Spine position is specified by a 2 tuple of (position type, amount). 
The position types are:

outward : place the spine out from the data area by the specified number of points. (Negative values specify placing the spine inward.)
axes : place the spine at the specified Axes coordinate (from 0.0-1.0).
data : place the spine at the specified data coordinate.
Additionally, shorthand notations define a special positions:

center -> (axes,0.5)
zero -> (data, 0.0)'''

ax = plt.gca()
ax.spines['right'].set_color('none') # make it colorless
ax.spines['top'].set_color('none') # make it colorless
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('axes',0.5))

ax.xaxis.set_label_coords(1, 0.5)
ax.yaxis.set_label_coords(0.5, -0.1)

# Save figure using 72 dots per inch
# plt.savefig("/nfs/science/figures/exercice_2.png",dpi=72)

plt.tight_layout() # use when your axis labels and title get clipped, will also adjust spacing between subplots to minimize the overlaps.
```

```python
ax=plt.subplot(211)          # the first subplot in the first figure
plt.plot([1, 2, 3],'r--')    # list of y values. x values will be generated randomly
plt.title("Subplot I", fontsize=16, fontweight='bold')     

plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6],'k')
plt.title("Subplot II", fontsize=16, fontweight='bold')     

#Annotate first plot
ax.annotate('This point', xy=(1,2))
# ax.annotate('This point', xy=(1,2), xytext=(1.5,2.5), arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)

# To switch off ticks/tick labels
from matplotlib.ticker import NullFormatter, NullLocator
ax.xaxis.set_major_formatter(NullFormatter())
ax.xaxis.set_major_locator(NullLocator())

ax.yaxis.set_major_formatter(NullFormatter())

plt.tight_layout()       
# plt.axis('off')
```

```python
# A simple bar chart
years=np.asarray([2012,2014,2016], dtype=int)

values = ([2, 5, 9])
values2 = [3,6,10]

plt.bar(years - 0.2, values, align='center',label='val1', width=0.4) # bar(x,y)
plt.xlabel('years', fontweight='bold')
plt.xticks(years, years)
plt.yticks(values)

plt.bar(years + 0.2, values2, color='orange', align='center', label='val2', width=0.4) # bar(x,y)

#use barh for horizontal bars
```

```python
# Stacked column chart
years=np.asarray([2012,2014,2016], dtype=int)

values = ([2, 5, 9])
values2 = [3,6,10]

plt.bar(years, values, color='r') 
plt.bar(years, values2, color='b', bottom=values) 
```
